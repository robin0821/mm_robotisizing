```python
import pandas as pd
import xlwings as xw
import datetime as dt
```

```python
file_prefix = dt.date.today().strftime('%y%m%d')
crm = pd.read_excel('C:\EY\Pipeline\Leadership\Account_List_Full.xlsx', sheetname='RevPlanExtract')
crm = crm.fillna("N.A.")
```

```python
crm.head()
```

```python
wb_mm = xw.Book(r'C:\\EY\\Pipeline\\MM\\' + file_prefix + "-MM_Intelligence_V01.xlsx")
sht_mm = wb_mm.sheets['Report']
```

```python
for i in range(2, 220):
    company = sht_mm.range('N' + str(i)).value
    company = str(company).lower()
    
    for l in crm['Client/Target Name'].index:
        full_name = crm['Client/Target Name'][l]
        full_name = str(full_name).lower()
        gcsp = crm['CSP'][l]
        gcsp = str(gcsp)
        tascl = crm['TAS Lead'][l]
        tascl = str(tascl)
        channel = crm['Channel'][l]
        channel = str(channel)
    
        if company not in full_name:
            continue
        else:
            sht_mm.range('P' + str(i)).value = channel # get channel No.
            sht_mm.range('Q' + str(i)).value = gcsp # get gcsp name
            sht_mm.range('R' + str(i)).value = tascl # get tascl name
            break
        l + 1
    i + 1
    print(i)            

wb_mm.save()
```
